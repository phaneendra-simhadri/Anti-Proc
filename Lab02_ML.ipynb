```python
import numpy as np
import pandas as pd

# - Each customer purchase is represented as a vector.
# - There are 10 customers, so there are 10 observation vectors.

# Load
df = pd.read_excel("/content/Lab Session Data.xlsx", sheet_name="Purchase data")

X = df[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']].values

# Rank tells us how many columns are linearly independent.
rank = np.linalg.matrix_rank(X)
print("Rank of Feature Matrix:", rank)

# output vector
y = df['Payment (Rs)'].values

# Since X is not a square matrix, its inverse cannot be computed.
X_pinv = np.linalg.pinv(X)

# Compute the cost of each product using:
cost = X_pinv @ y

#cost of each product
print(f"Candy        : ₹{cost[0]}")
print(f"Mango (1 Kg) : ₹{cost[1]}")
print(f"Milk Packet  : ₹{cost[2]}")
```
